```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('until-totaltime.csv')
```

```python
df
```

```python
df.head()
```

```python
starting_station=df["from_station_name"].to_numpy(dtype=None,copy=False)
```

```python
destination=df["to_station_name"].to_numpy(dtype=None,copy=False)
```

```python
starting_station_names, count1 = np.unique(starting_station, return_counts=True)
destination_names, count2 = np.unique(starting_station, return_counts=True)
```

```python
data = {'S.station':starting_station_names,
        'Frequency':count1}
```

```python
starting_frame=pd.DataFrame(data)
```

```python
starting_frame=starting_frame.sort_values(by='Frequency',ascending=False)
starting_frame=starting_frame.head(20)
```

```python
starting_frame
```

```python
data1 = {'Destination':destination_names,
        'Frequency':count2}
```

```python
destination_frame=pd.DataFrame(data1)
```

```python
destination_frame=destination_frame.sort_values(by='Frequency',ascending=False)
destination_frame=destination_frame.head(20)
```

```python
destination_frame
```

```python
#OBSERVATION : TO AND FRO SAME NUMBER OF TRAINS FOR A PARTICULAR STATION (except 1 or 2 deviation which is not visible in top 20)
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
y_pos = np.arange(len(destination_frame['Destination']))

ax.barh(y_pos,destination_frame['Frequency'], align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(destination_frame['Destination'])
ax.invert_yaxis()  
ax.set_xlabel('Frequency')
ax.set_title('Top 20 busiest stations')

plt.show()
```
